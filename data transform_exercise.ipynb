```python
fire_df = spark.read.table("dev.spark_db.sf_fire_calls")
```

```python
# distinct types of calls were made to teh fire dept?
result_df = (
    fire_df.selectExpr("count(distinct(CallType)) as distinct_call_type_count") )
```

```python
result_df.show()
```

```python
%sql
  select CallNumber, Delay
  from dev.spark_db.sf_fire_calls
  where Delay > 5
```

```python
result_df = (
    fire_df.select("CallNumber","Delay")
    .where("Delay >  5")
)

result_df.display()
```

```python
%sql
 select CallType, count(*) as count
  from dev.spark_db.sf_fire_calls
  where CallType is not null
  group by CallType
  order by count desc
```

```python
result_df = (
    fire_df.where("CallType is not null")
    .groupBy("CallType")
    .count()
    .orderBy("count",ascending= False)
)
result_df.display()
```

```python
%sql
  select distinct Neighborhood, Zipcode
  from dev.spark_db.sf_fire_calls
  where Zipcode== 94102 or Zipcode == 94103
```

```python
result_df = (
    fire_df.select("Neighborhood","Zipcode")
    .where("Zipcode ==94102 or Zipcode == 94103")
    .distinct()
)

result_df.display()
```

```python
%sql
 select sum(NumAlarms), avg(Delay), min(Delay), max(Delay)
  from dev.spark_db.sf_fire_calls
```

```python

(fire_df.selectExpr("sum(NumAlarms), avg(Delay), min(Delay), max(Delay)").show())


```

```python
%sql
select distinct year(CallDate) as year_num
  from dev.spark_db.sf_fire_calls
  order by year_num
```

```python
result_df =(
    fire_df.selectExpr("CallDate as year_num")
    .distinct()
    .orderBy("year_num")
)

result_df.show()
```

```python
%sql
select weekofyear(CallDate) as week_year, count(*) as count
  from dev.spark_db.sf_fire_calls
  where year(CallDate) == 2018
  group by week_year
  order by count desc
```

```python
(fire_df.selectExpr("weekofyear(CallDate) as week_year")
    .where("year(CallDate) == 2018")
    .groupBy('week_year')
    .count()
    .orderBy('count', ascending=False)
    .display()
)
```
