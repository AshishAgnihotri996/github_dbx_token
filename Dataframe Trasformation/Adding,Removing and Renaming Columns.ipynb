```python
schema="""
    id int,
    name string,
    age short,
    salary double
"""
```

```python
data_list = [(100,"Ashish",28,45000),
             (101,"Milind",29,33000),
             (102,"Ashwin",28,28000),
             ]
sample_df = spark.createDataFrame(data=data_list,schema=schema)
# .toDF("id","name","age","salary")
```

```python
sample_df.display()
```

add column in dataframe

1. increment 10% of salary upto 3000 maximum increment
2. revised salary: salary + increment

```python
from pyspark.sql.functions import expr
salary_df= (
        sample_df.withColumn
            ("increment",expr("case when salary >30000 then 3000 else salary * 10/100 end"))
        .withColumn
            ("revised_salary",expr("salary+increment"))
)
            
salary_df.display()
```

add a batch number uuid column in the dataframe

```python
salary_df.display()
```

add batch id to the column



```python
import uuid
from pyspark.sql.functions import lit

batch_id = str(uuid.uuid4())

salary_batch_df = salary_df.withColumn("batch_id",lit(batch_id))

salary_batch_df.display()
```

rename the df  
1. increment - annual_increment
2. salary - incremented_salary

```python
new_salary_df = (
    salary_batch_df.withColumnsRenamed({
        "increment" : "annual_increment",
        "salary" : "incremented_salary"
    })
)

new_salary_df.display()
```

remove the column 
1. age
2. annual_increment

```python
rm_salary_df = new_salary_df.drop("age","annual_increment")
rm_salary_df.display()
```
