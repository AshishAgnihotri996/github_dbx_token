```python
schema="""
    id int,
    name string,
    age short,
    salary double
"""
```

```python
data_list = [(100,"Ashish",28,45000),
             (101,"Milind",29,33000),
             (102,"Ashwin",28,28000),
             ]
sample_df = spark.createDataFrame(data=data_list,schema=schema)
# .toDF("id","name","age","salary")
```

```python
sample_df.display()
```

add column in dataframe

1. increment 10% of salary upto 3000 maximum increment
2. revised salary: salary + increment

```python
from pyspark.sql.functions import expr
salary_df= (
        sample_df.withColumn
            ("increment",expr("case when salary >30000 then 3000 else salary * 10/100 end"))
        .withColumn
            ("revised_salary",expr("salary+increment"))
)
            
salary_df.display()
```
